```python
import pandas as pd
import cv2

img_path = 'houses.jpg'
csv_path = 'colors.csv'
```

```python
index = ['color','color_name','hex','r','g','b']
df = pd.read_csv('colors.csv',names=index, header=None)
df.head(5)
```

```python

#Reading and resizing the image
img = cv2.imread(img_path)
img = cv2.resize(img,(800,600))

clicked = False
r = g = b = xpos = ypos = 0


def get_color_name(R,G,B):
    minimum = 1000
    for i in range(len(df)):
        d = abs(R - int(df.loc[i,'r'])) +  abs(G - int(df.loc[i,'g'])) +  abs(B - int(df.loc[i,'b']))
        if d<=minimum:
            minimum =d
            cname = df.loc[i , 'color_name']
    return cname


# This function is defined for opencv , we are binding this function with the mouse buttons
def draw_function(event,x,y,flags,params):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        # to make the variable global
        global clicked, r,g,b,xpos,ypos
        clicked = True
        xpos = x
        ypos = y
        b,g,r = img[y,x]
        b = int(b)
        g = int(g)
        r = int(r)
        
# open cv stores the colors in b g r format instead of r g b so we have to extract it    
    


#For the window to open we use use waitkey until it is closed
cv2.namedWindow('image')
# event function
cv2.setMouseCallback('image',draw_function) # when we left double click we get the co-ordinates of the image

while True:
    cv2.imshow('image',img)
    if clicked == True:
        cv2.rectangle(img ,(20,20),(600,60),(b,g,r),-1)
        text = get_color_name(r,g,b) + ' R='+ str(r) + ' G='+ str(g) + ' B='+ str(b)
        cv2.putText(img,text,(50,50),2,0.8,(255,255,255),2,cv2.LINE_AA)
        if r+g+b >= 600:
             cv2.putText(img,text,(50,50),2,0.8,(0,0,0),2,cv2.LINE_AA)
                
# if the user doesn't click in 20 seconds or presses escape key

    if cv2.waitKey(20) & 0xFF == 27:
        break
            
cv2.destroyAllWindows()    

#cv2.imshow('image',img)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
